```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset("mpg")

dataset.head()
```

```python
data_origin = dataset["origin"].value_counts().to_dict()

data_pie = data_origin.values()
label_pie = data_origin.keys()
colors = sns.color_palette('pastel')[0:5]
```

```python
plt.pie( x = data_pie,
        labels = label_pie,
        colors = colors,
        autopct = '%.0f%%')
plt.title("Origin")
plt.show()
```

```python
sns.histplot(data = dataset, x = "mpg", bins = 30)
plt.show()
```

```python
sns.boxplot(data = dataset, x = "mpg")
plt.show()
```

```python
sns.barplot(data = dataset, x = "origin", y = "mpg")
plt.show()
```

```python
sns.lineplot (data = dataset,
              x = "model_year",
              y = "mpg",
              hue = "origin")
```

```python
sns.scatterplot(data = dataset,
                x = "displacement",
                y = "mpg")
```
