## Identify unique station names

### Subtask:
Get the unique values from the 'stasiun' column of the `dt_ISPU` DataFrame.


**Reasoning**:
Access the 'stasiun' column and get the unique values using the `.unique()` method.



```python
unique_stations = dt_ISPU['stasiun'].unique()
display(unique_stations)
```

## Group similar names

### Subtask:
Implement a fuzzy matching algorithm to group similar station names based on a similarity threshold.


**Reasoning**:
Implement the fuzzy matching algorithm to group similar station names based on a similarity threshold and display the result.



```python
def group_similar_names(names, threshold):
    grouped_names = {}
    for name in names:
        found_match = False
        for grouped_name in grouped_names.keys():
            similarity = process.extractOne(name, [grouped_name])[1]
            if similarity >= threshold:
                grouped_names[grouped_name].append(name)
                found_match = True
                break
        if not found_match:
            grouped_names[name] = [name]
    return grouped_names

grouped_stations = group_similar_names(unique_stations, 85) # Using a threshold of 85
display(grouped_stations)
```

## Standardize names

### Subtask:
Replace the original station names in the 'stasiun' column with the chosen standardized name for each group.


**Reasoning**:
Iterate through the grouped_stations dictionary, create a mapping, and replace the station names in the DataFrame.



```python
station_mapping = {}
for standardized_name, original_names in grouped_stations.items():
    for original_name in original_names:
        station_mapping[original_name] = standardized_name

dt_ISPU['stasiun'] = dt_ISPU['stasiun'].replace(station_mapping)
display(dt_ISPU['stasiun'].unique())
```

## Verify changes

Display the unique values in the 'stasiun' column to confirm the standardization.


**Reasoning**:
